```python
import geopandas as gpd
 
# Load the shapefile
zones = gpd.read_file("taxi_zones/taxi_zones.shp")
print(zones.head())
 
# Compute centroids
zones['longitude'] = zones['geometry'].centroid.x
zones['latitude'] = zones['geometry'].centroid.y
```

```python
# compute the distance between two points
from geopy.distance import geodesic
def haversine(coord1, coord2):
    """
    Calculate the great circle distance in kilometers between two points
    on the earth specified in decimal degrees
    """
    # convert decimal degrees to radians
    coord1 = (coord1[0], coord1[1])
    coord2 = (coord2[0], coord2[1])
    
    # haversine formula
    from math import radians, sin, cos, sqrt, atan2
    R = 6371.0  # Radius of the Earth in kilometers
    dlon = radians(coord2[1] - coord1[1])
    dlat = radians(coord2[0] - coord1[0])
    a = sin(dlat/2)**2 + cos(radians(coord1[0])) * cos(radians(coord2[0])) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c  # Distance in kilometers
    # convert to miles
    distance = distance * 0.621371
    return distance

# extract locationId, longitude, and latitude
locations = zones[['OBJECTID', 'longitude', 'latitude']]
# rename objectId to locationId
locations = locations.rename(columns={'OBJECTID': 'locationid'})

# compute the distance between each pair of locations
from itertools import combinations
import numpy as np

def compute_distances(locations):
    distances = {}
    for loc1, loc2 in combinations(locations.iterrows(), 2):
        _, row1 = loc1
        _, row2 = loc2
        coord1 = (row1['latitude'], row1['longitude'])
        coord2 = (row2['latitude'], row2['longitude'])
        distance = haversine(coord1, coord2)
        distances[(row1['locationid'], row2['locationid'])] = distance
    return distances
# compute distances
distances = compute_distances(locations)
# convert distances to dataframe
distances_df = pd.DataFrame(distances.items(), columns=['pair', 'distance'])
# split pair into two columns
distances_df[['loc1', 'loc2']] = pd.DataFrame(distances_df['pair'].tolist(), index=distances_df.index)
# drop pair column
distances_df = distances_df.drop(columns=['pair'])
```

```python
loc_ids = range(1, 264)

# create a 2D matrix with all pairs of locations
distances_matrix = np.zeros((len(loc_ids), len(loc_ids)))

# fill the matrix with distances
for loc1, loc2 in combinations(loc_ids, 2):
    d = distances.get((loc1, loc2), None)
    if d is None:
        d = distances.get((loc2, loc1), None)

    distances_matrix[loc1-1, loc2-1] = d
    distances_matrix[loc2-1, loc1-1] = d
# fill the diagonal with zeros
for i in range(len(loc_ids)):
    distances_matrix[i, i] = 0
# convert to dataframe
distances_df = pd.DataFrame(distances_matrix, index=loc_ids, columns=loc_ids)
# save to csv
distances_df.to_csv('distances.csv', index=False)
```

```python
# create xgboost model to predict trip distance from pickup and dropoff locations
from sklearn.model_selection import train_test_split

# split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df[['PULocationID', 'DOLocationID']], df['trip_distance'], test_size=0.2, random_state=random_seed)
# create a xgboost model
model = xgb.XGBRegressor(objective='reg:squarederror', random_state=random_seed)
# fit the model
model.fit(X_train, y_train)
# predict the trip distance
y_pred = model.predict(X_test)
# evaluate the model
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Mean Squared Error:', mse)
print('R2 Score:', r2)
# plot the predicted vs actual trip distance
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Trip Distance')
plt.ylabel('Predicted Trip Distance')
plt.title('Predicted vs Actual Trip Distance')
plt.show()
# save the model
model.save_model('trip_distance_model.json')
```
